```python
d['date'] = pd.to_datetime(d['date'], dayfirst=True, errors='coerce')
d['date'] = d['date'].dt.strftime('%d-%m-%Y')
```

```python
d
```

```python
d.isnull().sum()
```

```python
d['transaction_type'] = d['transaction_type'].fillna(d['transaction_type'].mode()[0])
```

```python
d.isnull().sum()
```

```python
x=d[d['item_price'].isnull()]
y=d[~d['item_price'].isnull()]
```

```python
x
```

```python
y
```

```python
d.groupby('item_name')['item_price'].mean()
```

```python
def ad(x):
  if x=='Aalopuri':
    return 20
  elif x=='Cold coffee':
    return 40
  elif x=='Frankie':
    return 50
  elif x=='Sandwich':
    return 60
  elif x=='Sugarcane juice':
    return 25
  elif x=='Vadapav':
    return 20
  else:
    return 20
```

```python
x['item_price']=x['item_name'].apply(ad)
```

```python
x
```

```python
df=pd.concat((x,y),ignore_index=True)
df
```

```python
df.isnull().sum()
```

```python
x=df[df['item_type'].isnull()]
y=df[~df['item_type'].isnull()]
```

```python
x
```

```python
y
```

```python
def da(x):
  if x=='Aalopuri':
    return 'Fastfood'
  elif x=='Cold coffee':
    return 'Beverages'
  elif x=='Frankie':
    return 'Fastfood'
  elif x=='Sandwich':
    return 'Fastfood'
  elif x=='Sugarcane juice':
    return 'Beverages'
  elif x=='Vadapav':
    return 'Fastfood'
  else:
    return 'Fastfood'
```

```python
x['item_type']=x['item_name'].apply(da)
```

```python
x
```

```python
final=pd.concat((x,y),ignore_index=True)
final
```

```python
final.isnull().sum()
```

```python
final.to_csv('Sales cleaned_data.csv', index=False)
files.download('Sales cleaned_data.csv')
```
